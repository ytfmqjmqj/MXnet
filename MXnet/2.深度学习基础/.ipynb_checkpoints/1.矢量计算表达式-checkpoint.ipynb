```python
from mxnet import nd
from time import time
```

```python
a = nd.ones(shape = 1000)
b = nd.ones(shape = 1000)
```

## 将这两个向量元素逐一做标量加法

```python
start = time()
c = nd.zeros(shape = 1000)
for i in range(1000):
    c[i] = a[i] + b[i]
time() - start
```

## 直接做矢量加法

```python
start = time()
d = a + b
time() - start
```
